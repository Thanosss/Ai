### Unsloth

```python
from unsloth import FastLanguageModel  # ‰ªé unsloth Â∫ì‰∏≠ÂØºÂÖ• FastLanguageModel
import torch  # ÂØºÂÖ• PyTorch Â∫ì

max_seq_length = 2048  # ÊúÄÂ§ßÂ∫èÂàóÈïøÂ∫¶ÔºåÂèØ‰ª•ËÆæÁΩÆ‰ªªÊÑèÂÄºÔºåÂÜÖÈÉ®Ëá™Âä®ÊîØÊåÅ RoPE Êâ©Â±ï
dtype = None  # Êï∞ÊçÆÁ±ªÂûãÔºåËÆæÁΩÆ‰∏∫ None Ë°®Á§∫Ëá™Âä®Ê£ÄÊµã„ÄÇTesla T4 Âíå V100 ‰ΩøÁî® Float16ÔºåAmpere+ ‰ΩøÁî® Bfloat16
load_in_4bit = True  # ÊòØÂê¶‰ΩøÁî® 4bit ÈáèÂåñ‰ª•ÂáèÂ∞ëÂÜÖÂ≠ò‰ΩøÁî®ÔºåÂèØ‰ª•ËÆæÁΩÆ‰∏∫ False

# ÊîØÊåÅ 4bit È¢ÑÈáèÂåñÊ®°ÂûãÔºåËøô‰∫õÊ®°Âûã‰∏ãËΩΩÈÄüÂ∫¶Êõ¥Âø´‰∏î‰∏ç‰ºöÂá∫Áé∞ÂÜÖÂ≠ò‰∏çË∂≥ÔºàOOMÔºâ
fourbit_models = [
    "unsloth/Meta-Llama-3.1-8B-bnb-4bit",      # Llama-3.1 Ê®°ÂûãÔºå15‰∏á‰∫øËÆ≠ÁªÉÊï∞ÊçÆÔºåÈÄüÂ∫¶ÊèêÈ´ò 2 ÂÄç
    "unsloth/Meta-Llama-3.1-8B-Instruct-bnb-4bit",
    "unsloth/Meta-Llama-3.1-70B-bnb-4bit",
    "unsloth/Meta-Llama-3.1-405B-bnb-4bit",    # 405B ÂèÇÊï∞Ê®°ÂûãÁöÑ 4bit ÁâàÊú¨
    "unsloth/Mistral-Nemo-Base-2407-bnb-4bit", # Êñ∞Áâà Mistral 12BÔºåÈÄüÂ∫¶ÊèêÈ´ò 2 ÂÄç
    "unsloth/Mistral-Nemo-Instruct-2407-bnb-4bit",
    "unsloth/mistral-7b-v0.3-bnb-4bit",        # Mistral v3ÔºåÈÄüÂ∫¶ÊèêÈ´ò 2 ÂÄç
    "unsloth/mistral-7b-instruct-v0.3-bnb-4bit",
    "unsloth/Phi-3.5-mini-instruct",           # Phi-3.5ÔºåÈÄüÂ∫¶ÊèêÈ´ò 2 ÂÄç
    "unsloth/Phi-3-medium-4k-instruct",
    "unsloth/gemma-2-9b-bnb-4bit",
    "unsloth/gemma-2-27b-bnb-4bit",            # Gemma Ê®°ÂûãÈÄüÂ∫¶ÊèêÈ´ò 2 ÂÄç
]  # Êõ¥Â§öÊ®°ÂûãËØ∑ËÆøÈóÆ https://huggingface.co/unsloth

# Âä†ËΩΩÈ¢ÑËÆ≠ÁªÉÊ®°ÂûãÂíåÂàÜËØçÂô®
model, tokenizer = FastLanguageModel.from_pretrained(
    # ÂèØ‰ª•‰ªé‰ª•‰∏ãÊ®°Âûã‰∏≠ÈÄâÊã©Ôºö
    # "unsloth/Qwen2.5-0.5B", "unsloth/Qwen2.5-1.5B", "unsloth/Qwen2.5-3B"
    # "unsloth/Qwen2.5-14B",  "unsloth/Qwen2.5-32B",  "unsloth/Qwen2.5-72B",
    # ËøòÊîØÊåÅÊâÄÊúâ Instruct ÁâàÊú¨‰ª•ÂèäÊï∞Â≠¶„ÄÅ‰ª£Á†ÅÁõ∏ÂÖ≥ÁâàÊú¨ÔºÅ
    model_name = "unsloth/Qwen2.5-7B",  # ÊåáÂÆöË¶ÅÂä†ËΩΩÁöÑÊ®°ÂûãÂêçÁß∞
    max_seq_length = max_seq_length,    # ËÆæÁΩÆÊúÄÂ§ßÂ∫èÂàóÈïøÂ∫¶
    dtype = dtype,                      # ËÆæÁΩÆÊï∞ÊçÆÁ±ªÂûã
    load_in_4bit = load_in_4bit,        # ÊòØÂê¶ÂêØÁî® 4bit ÈáèÂåñ
    # token = "hf_...", # Â¶ÇÊûú‰ΩøÁî®ÂèóÈôêÊ®°ÂûãÔºàÂ¶Ç Meta-Llama/Llama-2-7b-hfÔºâÔºåÈúÄË¶ÅÊèê‰æõËÆøÈóÆ‰ª§Áâå
)
```

We now add LoRA adapters so we only need to update 1 to 10% of all parameters!

```python
model = FastLanguageModel.get_peft_model(
    model,  # Âü∫Á°ÄÊ®°Âûã
    r = 16,  # LoRA ÁöÑÁß©ÔºàrankÔºâÔºåÂèØ‰ª•ÈÄâÊã©‰ªªÊÑèÂ§ß‰∫é 0 ÁöÑÊï∞ÂÄºÔºåÂª∫ËÆÆÂÄº‰∏∫ 8„ÄÅ16„ÄÅ32„ÄÅ64„ÄÅ128
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj"],  # ÊåáÂÆöÈúÄË¶ÅÂ∫îÁî® LoRA ÁöÑÊ®°Âùó
    lora_alpha = 16,  # LoRA ÁöÑÁº©ÊîæÂõ†Â≠êÔºàalphaÔºâ
    lora_dropout = 0,  # LoRA ÁöÑ dropout ÂÄºÔºåÊîØÊåÅ‰ªªÊÑèÂÄºÔºå‰ΩÜËÆæÁΩÆ‰∏∫ 0 Êó∂ÊÄßËÉΩÊúÄ‰Ω≥
    bias = "none",     # LoRA ÁöÑÂÅèÁΩÆËÆæÁΩÆÔºåÊîØÊåÅ‰ªªÊÑèÂÄºÔºå‰ΩÜËÆæÁΩÆ‰∏∫ "none" Êó∂ÊÄßËÉΩÊúÄ‰Ω≥
    # [Êñ∞ÁâπÊÄß] "unsloth" ‰ΩøÁî® 30% Êõ¥Â∞ëÁöÑÊòæÂ≠òÔºåÊîØÊåÅ 2 ÂÄçÊõ¥Â§ßÁöÑÊâπÈáèÂ§ßÂ∞è
    use_gradient_checkpointing = "unsloth",  # ÂêØÁî®Ê¢ØÂ∫¶Ê£ÄÊü•ÁÇπÔºåÊîØÊåÅ True Êàñ "unsloth"ÔºåÁî®‰∫éË∂ÖÈïø‰∏ä‰∏ãÊñá
    random_state = 3407,  # ÈöèÊú∫ÁßçÂ≠êÔºåÁî®‰∫éÁ°Æ‰øùÁªìÊûúÂèØÂ§çÁé∞
    use_rslora = False,   # ÊòØÂê¶‰ΩøÁî® Rank Stabilized LoRAÔºàÁß©Á®≥ÂÆö LoRAÔºâ
    loftq_config = None,  # ÊòØÂê¶‰ΩøÁî® LoftQÔºàÈáèÂåñÈÖçÁΩÆÔºâ
)
```

<a name="Data"></a>
### Data Prep
We now use the Alpaca dataset from [yahma](https://huggingface.co/datasets/yahma/alpaca-cleaned), which is a filtered version of 52K of the original [Alpaca dataset](https://crfm.stanford.edu/2023/03/13/alpaca.html). You can replace this code section with your own data prep.

**[NOTE]** To train only on completions (ignoring the user's input) read TRL's docs [here](https://huggingface.co/docs/trl/sft_trainer#train-on-completions-only).

**[NOTE]** Remember to add the **EOS_TOKEN** to the tokenized output!! Otherwise you'll get infinite generations!

If you want to use the `llama-3` template for ShareGPT datasets, try our conversational [notebook](https://colab.research.google.com/github/unslothai/notebooks/blob/main/nb/Llama3_(8B)-Alpaca.ipynb)

For text completions like novel writing, try this [notebook](https://colab.research.google.com/github/unslothai/notebooks/blob/main/nb/Mistral_(7B)-Text_Completion.ipynb).

```python
# ÂÆö‰πâ Alpaca Ê†ºÂºèÂåñÊ®°Êùø
alpaca_prompt = """‰ª•‰∏ãÊòØÊèèËø∞‰ªªÂä°ÁöÑÊåá‰ª§Ôºå‰∏éÊèê‰æõËøõ‰∏ÄÊ≠•‰∏ä‰∏ãÊñáÁöÑËæìÂÖ•ÈÖçÂØπ„ÄÇÂÜô‰∏Ä‰∏™ÈÄÇÂΩìÁöÑÂìçÂ∫îÊù•ÂÆåÊàêËØ∑Ê±Ç„ÄÇ

### Instruction:
{}

### Input:
{}

### Response:
{}"""

EOS_TOKEN = tokenizer.eos_token  # ÂøÖÈ°ªÊ∑ªÂä† EOS_TOKENÔºàÁªìÊùüÊ†áËÆ∞Ôºâ
def formatting_prompts_func(examples):
    instructions = examples["instruction"]  # ‰ªéÊï∞ÊçÆÈõÜ‰∏≠ÊèêÂèñÊåá‰ª§
    inputs       = examples["input"]        # ‰ªéÊï∞ÊçÆÈõÜ‰∏≠ÊèêÂèñËæìÂÖ•
    outputs      = examples["output"]       # ‰ªéÊï∞ÊçÆÈõÜ‰∏≠ÊèêÂèñËæìÂá∫
    texts = []
    for instruction, input, output in zip(instructions, inputs, outputs):
        # ÂøÖÈ°ªÊ∑ªÂä† EOS_TOKENÔºåÂê¶ÂàôÁîüÊàêËøáÁ®ãÂèØËÉΩ‰ºöÊó†ÈôêËøõË°å
        text = alpaca_prompt.format(instruction, input, output) + EOS_TOKEN
        texts.append(text)
    return { "text" : texts }  # ËøîÂõûÊ†ºÂºèÂåñÂêéÁöÑÊñáÊú¨
pass

# Âä†ËΩΩÊï∞ÊçÆÈõÜ
from datasets import load_dataset
dataset = load_dataset("Thanosx/RegulatoryStandards", split = "train")  # Âä†ËΩΩÊåáÂÆöÊï∞ÊçÆÈõÜ
dataset = dataset.map(formatting_prompts_func, batched = True)  # ÊâπÈáèÂ∫îÁî®Ê†ºÂºèÂåñÂáΩÊï∞
```

<a name="Train"></a>
### Train the model
Now let's use Huggingface TRL's `SFTTrainer`! More docs here: [TRL SFT docs](https://huggingface.co/docs/trl/sft_trainer). We do 60 steps to speed things up, but you can set `num_train_epochs=1` for a full run, and turn off `max_steps=None`. We also support TRL's `DPOTrainer`!

```python
from trl import SFTTrainer  # ‰ªé trl Â∫ì‰∏≠ÂØºÂÖ• SFTTrainerÔºàÂæÆË∞ÉÂ∑•ÂÖ∑Ôºâ
from transformers import TrainingArguments  # ‰ªé transformers Â∫ì‰∏≠ÂØºÂÖ•ËÆ≠ÁªÉÂèÇÊï∞Á±ª
from unsloth import is_bfloat16_supported  # ‰ªé unsloth Â∫ì‰∏≠ÂØºÂÖ•Âà§Êñ≠ÊòØÂê¶ÊîØÊåÅ bfloat16 ÁöÑÂáΩÊï∞

trainer = SFTTrainer(
    model = model,  # Ë¶ÅÂæÆË∞ÉÁöÑÊ®°Âûã
    tokenizer = tokenizer,  # ‰∏éÊ®°ÂûãÂåπÈÖçÁöÑÂàÜËØçÂô®
    train_dataset = dataset,  # ËÆ≠ÁªÉÊï∞ÊçÆÈõÜ
    dataset_text_field = "text",  # Êï∞ÊçÆÈõÜ‰∏≠ÂåÖÂê´ÊñáÊú¨ÁöÑÂ≠óÊÆµÂêçÁß∞
    max_seq_length = max_seq_length,  # ÊúÄÂ§ßÂ∫èÂàóÈïøÂ∫¶
    dataset_num_proc = 2,  # Êï∞ÊçÆÈõÜÂ§ÑÁêÜÊó∂‰ΩøÁî®ÁöÑËøõÁ®ãÊï∞
    packing = False,  # ÊòØÂê¶ÂêØÁî®Êï∞ÊçÆÊâìÂåÖÔºàÈíàÂØπÁü≠Â∫èÂàóÔºåÂêØÁî®ÂèØ‰ª•Âä†ÈÄüËÆ≠ÁªÉÊúÄÂ§ö 5 ÂÄçÔºâ
    args = TrainingArguments(
        per_device_train_batch_size = 2,  # ÊØè‰∏™ËÆæÂ§áÁöÑËÆ≠ÁªÉÊâπÈáèÂ§ßÂ∞è
        gradient_accumulation_steps = 4,  # Ê¢ØÂ∫¶Á¥ØËÆ°Ê≠•Êï∞ÔºàÁõ∏ÂΩì‰∫éÊúâÊïàÊâπÈáèÂ§ßÂ∞è‰∏∫ 2 * 4 = 8Ôºâ
        warmup_steps = 5,  # Â≠¶‰π†ÁéáÁöÑÈ¢ÑÁÉ≠Ê≠•Êï∞
        # num_train_epochs = 1, # ËÆæÁΩÆÂÆåÊï¥ËÆ≠ÁªÉËΩÆÊï∞ÔºåÂèØÂêØÁî®Ê≠§ÂèÇÊï∞
        max_steps = 60,  # ÊúÄÂ§ßËÆ≠ÁªÉÊ≠•Êï∞
        learning_rate = 2e-4,  # Â≠¶‰π†Áéá
        fp16 = not is_bfloat16_supported(),  # ÊòØÂê¶ÂêØÁî® FP16ÔºàÂ¶ÇÊûú‰∏çÊîØÊåÅ bfloat16ÔºåÂàôÂêØÁî® FP16Ôºâ
        bf16 = is_bfloat16_supported(),  # ÊòØÂê¶ÂêØÁî® bfloat16ÔºàÂà§Êñ≠Á°¨‰ª∂ÊòØÂê¶ÊîØÊåÅÔºâ
        logging_steps = 1,  # Êó•ÂøóËÆ∞ÂΩïÁöÑÊ≠•Êï∞Èó¥Èöî
        optim = "adamw_8bit",  # ‰ºòÂåñÂô®Á±ªÂûãÔºå‰ΩøÁî® 8-bit AdamW ‰ºòÂåñÂô®‰ª•ÂáèÂ∞ëÊòæÂ≠òÂç†Áî®
        weight_decay = 0.01,  # ÊùÉÈáçË°∞ÂáèÁ≥ªÊï∞
        lr_scheduler_type = "linear",  # Â≠¶‰π†ÁéáË∞ÉÂ∫¶Âô®Á±ªÂûãÔºåÁ∫øÊÄßË°∞Âáè
        seed = 3407,  # ÈöèÊú∫ÁßçÂ≠êÔºåÁî®‰∫éÁ°Æ‰øùÁªìÊûúÂèØÂ§çÁé∞
        output_dir = "outputs",  # ËæìÂá∫ÁõÆÂΩïÔºåÁî®‰∫é‰øùÂ≠òÊ®°ÂûãÂíåÊó•Âøó
        report_to = "none",  # Êó•ÂøóÊä•ÂëäÊñπÂºèÔºàÂ¶Ç WandB Á≠âÔºâÔºåËÆæÁΩÆ‰∏∫ "none" Ë°®Á§∫‰∏ç‰ΩøÁî®
    ),
)
```

```python
# @title ÊòæÁ§∫ÂΩìÂâçÊòæÂ≠òÁä∂ÊÄÅ
gpu_stats = torch.cuda.get_device_properties(0)  # Ëé∑Âèñ GPU ÁöÑÂ±ûÊÄßÔºàËÆæÂ§á 0Ôºâ
start_gpu_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)  # Ëé∑ÂèñÂ∑≤‰øùÁïôÁöÑÊúÄÂ§ßÊòæÂ≠òÔºåÂπ∂ËΩ¨Êç¢‰∏∫ GB
max_memory = round(gpu_stats.total_memory / 1024 / 1024 / 1024, 3)  # Ëé∑Âèñ GPU ÁöÑÊÄªÊòæÂ≠òÔºåÂπ∂ËΩ¨Êç¢‰∏∫ GB
print(f"GPU = {gpu_stats.name}. ÊúÄÂ§ßÊòæÂ≠ò = {max_memory} GB.")  # ÊâìÂç∞ GPU ÂêçÁß∞ÂíåÊÄªÊòæÂ≠ò
print(f"{start_gpu_memory} GB ÁöÑÊòæÂ≠òÂ∑≤Ë¢´‰øùÁïô„ÄÇ")  # ÊâìÂç∞Â∑≤‰øùÁïôÁöÑÊòæÂ≠ò
```

```python
trainer_stats = trainer.train()
```

```python
# @title ÊòæÁ§∫ÊúÄÁªàÁöÑÊòæÂ≠òÂíåÊó∂Èó¥ÁªüËÆ°‰ø°ÊÅØ
used_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)  # Ëé∑ÂèñÂ≥∞ÂÄº‰øùÁïôÊòæÂ≠òÔºåÂπ∂ËΩ¨Êç¢‰∏∫ GB
used_memory_for_lora = round(used_memory - start_gpu_memory, 3)  # ËÆ°ÁÆóÂæÆË∞ÉËøáÁ®ã‰∏≠‰ΩøÁî®ÁöÑÊòæÂ≠òÔºàLoRA ÂæÆË∞ÉÈÉ®ÂàÜÔºâ
used_percentage = round(used_memory / max_memory * 100, 3)  # Â≥∞ÂÄº‰øùÁïôÊòæÂ≠òÂç†ÊÄªÊòæÂ≠òÁöÑÁôæÂàÜÊØî
lora_percentage = round(used_memory_for_lora / max_memory * 100, 3)  # ÂæÆË∞ÉÊòæÂ≠òÂç†ÊÄªÊòæÂ≠òÁöÑÁôæÂàÜÊØî

# ÊâìÂç∞ËÆ≠ÁªÉ‰ΩøÁî®ÁöÑÊó∂Èó¥ÔºàÁßíÔºâ
print(f"{trainer_stats.metrics['train_runtime']} ÁßíÁî®‰∫éËÆ≠ÁªÉ„ÄÇ")

# ÊâìÂç∞ËÆ≠ÁªÉ‰ΩøÁî®ÁöÑÊó∂Èó¥ÔºàÂàÜÈíüÔºâ
print(f"{round(trainer_stats.metrics['train_runtime']/60, 2)} ÂàÜÈíüÁî®‰∫éËÆ≠ÁªÉ„ÄÇ")

# ÊâìÂç∞Â≥∞ÂÄº‰øùÁïôÊòæÂ≠òÔºàGBÔºâ
print(f"Â≥∞ÂÄº‰øùÁïôÊòæÂ≠ò = {used_memory} GB„ÄÇ")

# ÊâìÂç∞ÂæÆË∞ÉËøáÁ®ã‰∏≠‰ΩøÁî®ÁöÑÊòæÂ≠òÔºàGBÔºâ
print(f"ÂæÆË∞ÉËøáÁ®ã‰∏≠‰ΩøÁî®ÁöÑÂ≥∞ÂÄºÊòæÂ≠ò = {used_memory_for_lora} GB„ÄÇ")

# ÊâìÂç∞Â≥∞ÂÄº‰øùÁïôÊòæÂ≠òÂç†ÊÄªÊòæÂ≠òÁöÑÁôæÂàÜÊØî
print(f"Â≥∞ÂÄº‰øùÁïôÊòæÂ≠òÂç†ÊÄªÊòæÂ≠òÁöÑÁôæÂàÜÊØî = {used_percentage} %„ÄÇ")

# ÊâìÂç∞ÂæÆË∞ÉÊòæÂ≠òÂç†ÊÄªÊòæÂ≠òÁöÑÁôæÂàÜÊØî
print(f"ÂæÆË∞ÉËøáÁ®ã‰∏≠ÊòæÂ≠òÂç†ÊÄªÊòæÂ≠òÁöÑÁôæÂàÜÊØî = {lora_percentage} %„ÄÇ")
```

```python
# alpaca_prompt = ‰ªé‰∏äÊñπÂ§çÂà∂ÁöÑÊ®°Êùø
FastLanguageModel.for_inference(model)  # ÂêØÁî®ÂéüÁîüÁöÑ 2 ÂÄçÈÄüÊé®ÁêÜÊ®°Âºè
inputs = tokenizer(
[
    alpaca_prompt.format(
        "ÁªßÁª≠ÊñêÊ≥¢ÈÇ£Â•ëÊï∞Âàó„ÄÇ",  # Êåá‰ª§
        "1, 1, 2, 3, 5, 8",  # ËæìÂÖ•
        "",  # ËæìÂá∫ - ÁïôÁ©∫‰ª•ËøõË°åÁîüÊàêÔºÅ
    )
], return_tensors="pt").to("cuda")  # Â∞ÜËæìÂÖ•ËΩ¨Êç¢‰∏∫Âº†ÈáèÂπ∂ËΩ¨ÁßªÂà∞ GPU

outputs = model.generate(**inputs, max_new_tokens=64, use_cache=True)  # ÁîüÊàêÊúÄÂ§ö 64 ‰∏™Êñ∞ token
tokenizer.batch_decode(outputs)  # Ëß£Á†ÅÁîüÊàêÁöÑËæìÂá∫
```

**[Ê≥®ÊÑè]** ËøôÂè™‰ºö‰øùÂ≠ò LoRA ÈÄÇÈÖçÂô®ÔºåËÄå‰∏ç‰ºö‰øùÂ≠òÂÆåÊï¥Ê®°Âûã„ÄÇÂ¶ÇÊûúÈúÄË¶Å‰øùÂ≠ò‰∏∫ 16bit Êàñ GGUF Ê†ºÂºèÔºåËØ∑Âêë‰∏ãÊªöÂä®Êü•ÁúãÔºÅ

```python
model.save_pretrained("lora_model")  # Êú¨Âú∞‰øùÂ≠òÊ®°Âûã
tokenizer.save_pretrained("lora_model")  # Êú¨Âú∞‰øùÂ≠òÂàÜËØçÂô®
# model.push_to_hub("your_name/lora_model", token="...")  # Âú®Á∫ø‰øùÂ≠òÊ®°ÂûãÂà∞ Hugging Face Hub
# tokenizer.push_to_hub("your_name/lora_model", token="...")  # Âú®Á∫ø‰øùÂ≠òÂàÜËØçÂô®Âà∞ Hugging Face Hub
```

Áé∞Âú®ÔºåÂ¶ÇÊûú‰Ω†ÊÉ≥Âä†ËΩΩÊàë‰ª¨ÂàöÂàö‰øùÂ≠òÁöÑ LoRA ÈÄÇÈÖçÂô®ËøõË°åÊé®ÁêÜÔºåÂè™ÈúÄÂ∞Ü `False` ËÆæÁΩÆ‰∏∫ `True`Ôºö

```python
if False:  # Â¶ÇÊûúÊù°‰ª∂‰∏∫ TrueÔºåÂàôÊâßË°å‰ª•‰∏ã‰ª£Á†ÅÔºàÂΩìÂâçÊòØ FalseÔºå‰ª£Á†Å‰∏ç‰ºöËøêË°åÔºâ
    from unsloth import FastLanguageModel  # ‰ªé unsloth Â∫ìÂØºÂÖ• FastLanguageModel
    model, tokenizer = FastLanguageModel.from_pretrained(
        model_name="lora_model",  # ‰Ω†Áî®‰∫éËÆ≠ÁªÉÁöÑÊ®°ÂûãÂêçÁß∞
        max_seq_length=max_seq_length,  # ÊúÄÂ§ßÂ∫èÂàóÈïøÂ∫¶
        dtype=dtype,  # Êï∞ÊçÆÁ±ªÂûãÔºà‰æãÂ¶Ç fp16„ÄÅfp32 Á≠âÔºâ
        load_in_4bit=load_in_4bit,  # ÊòØÂê¶‰ª• 4bit Ê†ºÂºèÂä†ËΩΩÊ®°ÂûãÔºàËäÇÁúÅÊòæÂ≠òÔºâ
    )
    FastLanguageModel.for_inference(model)  # ÂêØÁî®ÂéüÁîüÁöÑ 2 ÂÄçÈÄüÊé®ÁêÜÊ®°Âºè

# alpaca_prompt = ‰Ω†ÂøÖÈ°ª‰ªé‰∏äÊñπÂ§çÂà∂Ê®°ÊùøÔºÅ

inputs = tokenizer(
[
    alpaca_prompt.format(
        "Â∑¥ÈªéËëóÂêçÁöÑÈ´òÂ°îÊòØ‰ªÄ‰πàÔºü",  # Êåá‰ª§
        "",  # ËæìÂÖ•
        "",  # ËæìÂá∫ - ÁïôÁ©∫‰ª•ÁîüÊàêÂÜÖÂÆπÔºÅ
    )
], return_tensors="pt").to("cuda")  # Â∞ÜËæìÂÖ•ËΩ¨Êç¢‰∏∫ PyTorch Âº†ÈáèÔºåÂπ∂ÁßªÂä®Âà∞ GPU

from transformers import TextStreamer  # ‰ªé transformers Â∫ìÂØºÂÖ• TextStreamer
text_streamer = TextStreamer(tokenizer)  # ÂàõÂª∫‰∏Ä‰∏™ÊñáÊú¨ÊµÅÂØπË±°ÔºåÁî®‰∫éÈÄêÊ≠•ÊòæÁ§∫ÁîüÊàêÂÜÖÂÆπ
_ = model.generate(**inputs, streamer=text_streamer, max_new_tokens=128)  # ‰ΩøÁî®ÊµÅÂºèÁîüÊàêÔºåÊúÄÂ§öÁîüÊàê 128 ‰∏™Êñ∞ token
```

‰Ω†‰πüÂèØ‰ª•‰ΩøÁî® Hugging Face ÁöÑ `AutoModelForPeftCausalLM`„ÄÇ‰ªÖÂú®Êú™ÂÆâË£Ö `unsloth` ÁöÑÊÉÖÂÜµ‰∏ã‰ΩøÁî®Ê≠§ÊñπÊ≥ï„ÄÇ‰∏çËøáÔºåËøôÂèØËÉΩ‰ºöÈùûÂ∏∏ÊÖ¢ÔºåÂõ†‰∏∫‰∏çÊîØÊåÅ‰∏ãËΩΩ `4bit` Ê®°ÂûãÔºåËÄå‰ΩøÁî® Unsloth ÁöÑ**Êé®ÁêÜÈÄüÂ∫¶Âø´ 2 ÂÄç**„ÄÇ

```python
if False:  # Â¶ÇÊûúÊù°‰ª∂‰∏∫ TrueÔºåÂàôÊâßË°å‰ª•‰∏ã‰ª£Á†ÅÔºàÂΩìÂâçÊòØ FalseÔºå‰ª£Á†Å‰∏ç‰ºöËøêË°åÔºâ
    # ÊàëÂº∫ÁÉà‰∏çÊé®Ëçê‰ΩøÁî®ËøôÁßçÊñπÂºè - Â¶ÇÊûúÂèØËÉΩÔºåËØ∑‰ΩøÁî® Unsloth
    from peft import AutoPeftModelForCausalLM  # ‰ªé peft Â∫ìÂØºÂÖ• AutoPeftModelForCausalLM
    from transformers import AutoTokenizer  # ‰ªé transformers Â∫ìÂØºÂÖ• AutoTokenizer
    model = AutoPeftModelForCausalLM.from_pretrained(
        "lora_model",  # ‰Ω†Áî®‰∫éËÆ≠ÁªÉÁöÑÊ®°ÂûãÂêçÁß∞
        load_in_4bit=load_in_4bit,  # ÊòØÂê¶‰ª• 4bit Ê†ºÂºèÂä†ËΩΩÊ®°ÂûãÔºàËäÇÁúÅÊòæÂ≠òÔºâ
    )
    tokenizer = AutoTokenizer.from_pretrained("lora_model")  # Âä†ËΩΩÂàÜËØçÂô®
```

### ‰øùÂ≠ò‰∏∫ float16 Ê†ºÂºè‰ª•‰æõ VLLM ‰ΩøÁî®

Êàë‰ª¨ÊîØÊåÅÁõ¥Êé•Â∞ÜÊ®°Âûã‰øùÂ≠ò‰∏∫ `float16` Ê†ºÂºè„ÄÇÂèØ‰ª•ÈÄâÊã© `merged_16bit` ‰øùÂ≠ò‰∏∫ float16 Ê†ºÂºèÔºåÊàñËÄÖÈÄâÊã© `merged_4bit` ‰øùÂ≠ò‰∏∫ int4 Ê†ºÂºè„ÄÇÊ≠§Â§ñÔºåËøòÊîØÊåÅ‰ΩøÁî® `lora` ÈÄÇÈÖçÂô®‰Ωú‰∏∫Â§áÁî®ÈÄâÈ°π„ÄÇ

‰ΩøÁî® `push_to_hub_merged` ÊñπÊ≥ïÂèØ‰ª•Â∞ÜÊ®°Âûã‰∏ä‰º†Âà∞‰Ω†ÁöÑ Hugging Face Ë¥¶Êà∑ÔºÅËÆøÈóÆ [https://huggingface.co/settings/tokens](https://huggingface.co/settings/tokens) Ëé∑Âèñ‰Ω†ÁöÑ‰∏™‰∫∫ËÆøÈóÆ‰ª§Áâå„ÄÇ

```python
# ÂêàÂπ∂‰∏∫ 16bit Ê®°Âûã
if False:
    model.save_pretrained_merged("model", tokenizer, save_method="merged_16bit",)
if False:
    model.push_to_hub_merged("hf/model", tokenizer, save_method="merged_16bit", token="")

# ÂêàÂπ∂‰∏∫ 4bit Ê®°Âûã
if False:
    model.save_pretrained_merged("model", tokenizer, save_method="merged_4bit",)
if False:
    model.push_to_hub_merged("hf/model", tokenizer, save_method="merged_4bit", token="")

# ‰ªÖ‰øùÂ≠ò LoRA ÈÄÇÈÖçÂô®
if False:
    model.save_pretrained_merged("model", tokenizer, save_method="lora",)
if False:
    model.push_to_hub_merged("hf/model", tokenizer, save_method="lora", token="")
```
