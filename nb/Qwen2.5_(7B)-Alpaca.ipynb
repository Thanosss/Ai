### Unsloth

```python
from unsloth import FastLanguageModel  # ä» unsloth åº“ä¸­å¯¼å…¥ FastLanguageModel
import torch  # å¯¼å…¥ PyTorch åº“

max_seq_length = 2048  # æœ€å¤§åºåˆ—é•¿åº¦ï¼Œå¯ä»¥è®¾ç½®ä»»æ„å€¼ï¼Œå†…éƒ¨è‡ªåŠ¨æ”¯æŒ RoPE æ‰©å±•
dtype = None  # æ•°æ®ç±»å‹ï¼Œè®¾ç½®ä¸º None è¡¨ç¤ºè‡ªåŠ¨æ£€æµ‹ã€‚Tesla T4 å’Œ V100 ä½¿ç”¨ Float16ï¼ŒAmpere+ ä½¿ç”¨ Bfloat16
load_in_4bit = True  # æ˜¯å¦ä½¿ç”¨ 4bit é‡åŒ–ä»¥å‡å°‘å†…å­˜ä½¿ç”¨ï¼Œå¯ä»¥è®¾ç½®ä¸º False

# æ”¯æŒ 4bit é¢„é‡åŒ–æ¨¡å‹ï¼Œè¿™äº›æ¨¡å‹ä¸‹è½½é€Ÿåº¦æ›´å¿«ä¸”ä¸ä¼šå‡ºç°å†…å­˜ä¸è¶³ï¼ˆOOMï¼‰
fourbit_models = [
    "unsloth/Meta-Llama-3.1-8B-bnb-4bit",      # Llama-3.1 æ¨¡å‹ï¼Œ15ä¸‡äº¿è®­ç»ƒæ•°æ®ï¼Œé€Ÿåº¦æé«˜ 2 å€
    "unsloth/Meta-Llama-3.1-8B-Instruct-bnb-4bit",
    "unsloth/Meta-Llama-3.1-70B-bnb-4bit",
    "unsloth/Meta-Llama-3.1-405B-bnb-4bit",    # 405B å‚æ•°æ¨¡å‹çš„ 4bit ç‰ˆæœ¬
    "unsloth/Mistral-Nemo-Base-2407-bnb-4bit", # æ–°ç‰ˆ Mistral 12Bï¼Œé€Ÿåº¦æé«˜ 2 å€
    "unsloth/Mistral-Nemo-Instruct-2407-bnb-4bit",
    "unsloth/mistral-7b-v0.3-bnb-4bit",        # Mistral v3ï¼Œé€Ÿåº¦æé«˜ 2 å€
    "unsloth/mistral-7b-instruct-v0.3-bnb-4bit",
    "unsloth/Phi-3.5-mini-instruct",           # Phi-3.5ï¼Œé€Ÿåº¦æé«˜ 2 å€
    "unsloth/Phi-3-medium-4k-instruct",
    "unsloth/gemma-2-9b-bnb-4bit",
    "unsloth/gemma-2-27b-bnb-4bit",            # Gemma æ¨¡å‹é€Ÿåº¦æé«˜ 2 å€
]  # æ›´å¤šæ¨¡å‹è¯·è®¿é—® https://huggingface.co/unsloth

# åŠ è½½é¢„è®­ç»ƒæ¨¡å‹å’Œåˆ†è¯å™¨
model, tokenizer = FastLanguageModel.from_pretrained(
    # å¯ä»¥ä»ä»¥ä¸‹æ¨¡å‹ä¸­é€‰æ‹©ï¼š
    # "unsloth/Qwen2.5-0.5B", "unsloth/Qwen2.5-1.5B", "unsloth/Qwen2.5-3B"
    # "unsloth/Qwen2.5-14B",  "unsloth/Qwen2.5-32B",  "unsloth/Qwen2.5-72B",
    # è¿˜æ”¯æŒæ‰€æœ‰ Instruct ç‰ˆæœ¬ä»¥åŠæ•°å­¦ã€ä»£ç ç›¸å…³ç‰ˆæœ¬ï¼
    model_name = "unsloth/Qwen2.5-7B",  # æŒ‡å®šè¦åŠ è½½çš„æ¨¡å‹åç§°
    max_seq_length = max_seq_length,    # è®¾ç½®æœ€å¤§åºåˆ—é•¿åº¦
    dtype = dtype,                      # è®¾ç½®æ•°æ®ç±»å‹
    load_in_4bit = load_in_4bit,        # æ˜¯å¦å¯ç”¨ 4bit é‡åŒ–
    # token = "hf_...", # å¦‚æœä½¿ç”¨å—é™æ¨¡å‹ï¼ˆå¦‚ Meta-Llama/Llama-2-7b-hfï¼‰ï¼Œéœ€è¦æä¾›è®¿é—®ä»¤ç‰Œ
)
```

We now add LoRA adapters so we only need to update 1 to 10% of all parameters!

```python
model = FastLanguageModel.get_peft_model(
    model,  # åŸºç¡€æ¨¡å‹
    r = 16,  # LoRA çš„ç§©ï¼ˆrankï¼‰ï¼Œå¯ä»¥é€‰æ‹©ä»»æ„å¤§äº 0 çš„æ•°å€¼ï¼Œå»ºè®®å€¼ä¸º 8ã€16ã€32ã€64ã€128
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj"],  # æŒ‡å®šéœ€è¦åº”ç”¨ LoRA çš„æ¨¡å—
    lora_alpha = 16,  # LoRA çš„ç¼©æ”¾å› å­ï¼ˆalphaï¼‰
    lora_dropout = 0,  # LoRA çš„ dropout å€¼ï¼Œæ”¯æŒä»»æ„å€¼ï¼Œä½†è®¾ç½®ä¸º 0 æ—¶æ€§èƒ½æœ€ä½³
    bias = "none",     # LoRA çš„åç½®è®¾ç½®ï¼Œæ”¯æŒä»»æ„å€¼ï¼Œä½†è®¾ç½®ä¸º "none" æ—¶æ€§èƒ½æœ€ä½³
    # [æ–°ç‰¹æ€§] "unsloth" ä½¿ç”¨ 30% æ›´å°‘çš„æ˜¾å­˜ï¼Œæ”¯æŒ 2 å€æ›´å¤§çš„æ‰¹é‡å¤§å°
    use_gradient_checkpointing = "unsloth",  # å¯ç”¨æ¢¯åº¦æ£€æŸ¥ç‚¹ï¼Œæ”¯æŒ True æˆ– "unsloth"ï¼Œç”¨äºè¶…é•¿ä¸Šä¸‹æ–‡
    random_state = 3407,  # éšæœºç§å­ï¼Œç”¨äºç¡®ä¿ç»“æœå¯å¤ç°
    use_rslora = False,   # æ˜¯å¦ä½¿ç”¨ Rank Stabilized LoRAï¼ˆç§©ç¨³å®š LoRAï¼‰
    loftq_config = None,  # æ˜¯å¦ä½¿ç”¨ LoftQï¼ˆé‡åŒ–é…ç½®ï¼‰
)
```

<a name="Data"></a>
### Data Prep
We now use the Alpaca dataset from [yahma](https://huggingface.co/datasets/yahma/alpaca-cleaned), which is a filtered version of 52K of the original [Alpaca dataset](https://crfm.stanford.edu/2023/03/13/alpaca.html). You can replace this code section with your own data prep.

**[NOTE]** To train only on completions (ignoring the user's input) read TRL's docs [here](https://huggingface.co/docs/trl/sft_trainer#train-on-completions-only).

**[NOTE]** Remember to add the **EOS_TOKEN** to the tokenized output!! Otherwise you'll get infinite generations!

If you want to use the `llama-3` template for ShareGPT datasets, try our conversational [notebook](https://colab.research.google.com/github/unslothai/notebooks/blob/main/nb/Llama3_(8B)-Alpaca.ipynb)

For text completions like novel writing, try this [notebook](https://colab.research.google.com/github/unslothai/notebooks/blob/main/nb/Mistral_(7B)-Text_Completion.ipynb).

```python
# å®šä¹‰ Alpaca æ ¼å¼åŒ–æ¨¡æ¿
alpaca_prompt = """ä»¥ä¸‹æ˜¯æè¿°ä»»åŠ¡çš„æŒ‡ä»¤ï¼Œä¸æä¾›è¿›ä¸€æ­¥ä¸Šä¸‹æ–‡çš„è¾“å…¥é…å¯¹ã€‚å†™ä¸€ä¸ªé€‚å½“çš„å“åº”æ¥å®Œæˆè¯·æ±‚ã€‚

### Instruction:
{}

### Input:
{}

### Response:
{}"""

EOS_TOKEN = tokenizer.eos_token  # å¿…é¡»æ·»åŠ  EOS_TOKENï¼ˆç»“æŸæ ‡è®°ï¼‰
def formatting_prompts_func(examples):
    instructions = examples["instruction"]  # ä»æ•°æ®é›†ä¸­æå–æŒ‡ä»¤
    inputs       = examples["input"]        # ä»æ•°æ®é›†ä¸­æå–è¾“å…¥
    outputs      = examples["output"]       # ä»æ•°æ®é›†ä¸­æå–è¾“å‡º
    texts = []
    for instruction, input, output in zip(instructions, inputs, outputs):
        # å¿…é¡»æ·»åŠ  EOS_TOKENï¼Œå¦åˆ™ç”Ÿæˆè¿‡ç¨‹å¯èƒ½ä¼šæ— é™è¿›è¡Œ
        text = alpaca_prompt.format(instruction, input, output) + EOS_TOKEN
        texts.append(text)
    return { "text" : texts }  # è¿”å›æ ¼å¼åŒ–åçš„æ–‡æœ¬
pass

# åŠ è½½æ•°æ®é›†
from datasets import load_dataset
dataset = load_dataset("Thanosx/RegulatoryStandards", split = "train")  # åŠ è½½æŒ‡å®šæ•°æ®é›†
dataset = dataset.map(formatting_prompts_func, batched = True)  # æ‰¹é‡åº”ç”¨æ ¼å¼åŒ–å‡½æ•°
```

<a name="Train"></a>
### Train the model
Now let's use Huggingface TRL's `SFTTrainer`! More docs here: [TRL SFT docs](https://huggingface.co/docs/trl/sft_trainer). We do 60 steps to speed things up, but you can set `num_train_epochs=1` for a full run, and turn off `max_steps=None`. We also support TRL's `DPOTrainer`!

```python
from trl import SFTTrainer  # ä» trl åº“ä¸­å¯¼å…¥ SFTTrainerï¼ˆå¾®è°ƒå·¥å…·ï¼‰
from transformers import TrainingArguments  # ä» transformers åº“ä¸­å¯¼å…¥è®­ç»ƒå‚æ•°ç±»
from unsloth import is_bfloat16_supported  # ä» unsloth åº“ä¸­å¯¼å…¥åˆ¤æ–­æ˜¯å¦æ”¯æŒ bfloat16 çš„å‡½æ•°

trainer = SFTTrainer(
    model = model,  # è¦å¾®è°ƒçš„æ¨¡å‹
    tokenizer = tokenizer,  # ä¸æ¨¡å‹åŒ¹é…çš„åˆ†è¯å™¨
    train_dataset = dataset,  # è®­ç»ƒæ•°æ®é›†
    dataset_text_field = "text",  # æ•°æ®é›†ä¸­åŒ…å«æ–‡æœ¬çš„å­—æ®µåç§°
    max_seq_length = max_seq_length,  # æœ€å¤§åºåˆ—é•¿åº¦
    dataset_num_proc = 2,  # æ•°æ®é›†å¤„ç†æ—¶ä½¿ç”¨çš„è¿›ç¨‹æ•°
    packing = False,  # æ˜¯å¦å¯ç”¨æ•°æ®æ‰“åŒ…ï¼ˆé’ˆå¯¹çŸ­åºåˆ—ï¼Œå¯ç”¨å¯ä»¥åŠ é€Ÿè®­ç»ƒæœ€å¤š 5 å€ï¼‰
    args = TrainingArguments(
        per_device_train_batch_size = 2,  # æ¯ä¸ªè®¾å¤‡çš„è®­ç»ƒæ‰¹é‡å¤§å°
        gradient_accumulation_steps = 4,  # æ¢¯åº¦ç´¯è®¡æ­¥æ•°ï¼ˆç›¸å½“äºæœ‰æ•ˆæ‰¹é‡å¤§å°ä¸º 2 * 4 = 8ï¼‰
        warmup_steps = 5,  # å­¦ä¹ ç‡çš„é¢„çƒ­æ­¥æ•°
        # num_train_epochs = 1, # è®¾ç½®å®Œæ•´è®­ç»ƒè½®æ•°ï¼Œå¯å¯ç”¨æ­¤å‚æ•°
        max_steps = 60,  # æœ€å¤§è®­ç»ƒæ­¥æ•°
        learning_rate = 2e-4,  # å­¦ä¹ ç‡
        fp16 = not is_bfloat16_supported(),  # æ˜¯å¦å¯ç”¨ FP16ï¼ˆå¦‚æœä¸æ”¯æŒ bfloat16ï¼Œåˆ™å¯ç”¨ FP16ï¼‰
        bf16 = is_bfloat16_supported(),  # æ˜¯å¦å¯ç”¨ bfloat16ï¼ˆåˆ¤æ–­ç¡¬ä»¶æ˜¯å¦æ”¯æŒï¼‰
        logging_steps = 1,  # æ—¥å¿—è®°å½•çš„æ­¥æ•°é—´éš”
        optim = "adamw_8bit",  # ä¼˜åŒ–å™¨ç±»å‹ï¼Œä½¿ç”¨ 8-bit AdamW ä¼˜åŒ–å™¨ä»¥å‡å°‘æ˜¾å­˜å ç”¨
        weight_decay = 0.01,  # æƒé‡è¡°å‡ç³»æ•°
        lr_scheduler_type = "linear",  # å­¦ä¹ ç‡è°ƒåº¦å™¨ç±»å‹ï¼Œçº¿æ€§è¡°å‡
        seed = 3407,  # éšæœºç§å­ï¼Œç”¨äºç¡®ä¿ç»“æœå¯å¤ç°
        output_dir = "outputs",  # è¾“å‡ºç›®å½•ï¼Œç”¨äºä¿å­˜æ¨¡å‹å’Œæ—¥å¿—
        report_to = "none",  # æ—¥å¿—æŠ¥å‘Šæ–¹å¼ï¼ˆå¦‚ WandB ç­‰ï¼‰ï¼Œè®¾ç½®ä¸º "none" è¡¨ç¤ºä¸ä½¿ç”¨
    ),
)
```

```python
# @title æ˜¾ç¤ºå½“å‰æ˜¾å­˜çŠ¶æ€
gpu_stats = torch.cuda.get_device_properties(0)  # è·å– GPU çš„å±æ€§ï¼ˆè®¾å¤‡ 0ï¼‰
start_gpu_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)  # è·å–å·²ä¿ç•™çš„æœ€å¤§æ˜¾å­˜ï¼Œå¹¶è½¬æ¢ä¸º GB
max_memory = round(gpu_stats.total_memory / 1024 / 1024 / 1024, 3)  # è·å– GPU çš„æ€»æ˜¾å­˜ï¼Œå¹¶è½¬æ¢ä¸º GB
print(f"GPU = {gpu_stats.name}. æœ€å¤§æ˜¾å­˜ = {max_memory} GB.")  # æ‰“å° GPU åç§°å’Œæ€»æ˜¾å­˜
print(f"{start_gpu_memory} GB çš„æ˜¾å­˜å·²è¢«ä¿ç•™ã€‚")  # æ‰“å°å·²ä¿ç•™çš„æ˜¾å­˜
```

```python
trainer_stats = trainer.train()
```

```python
# @title æ˜¾ç¤ºæœ€ç»ˆçš„æ˜¾å­˜å’Œæ—¶é—´ç»Ÿè®¡ä¿¡æ¯
used_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)  # è·å–å³°å€¼ä¿ç•™æ˜¾å­˜ï¼Œå¹¶è½¬æ¢ä¸º GB
used_memory_for_lora = round(used_memory - start_gpu_memory, 3)  # è®¡ç®—å¾®è°ƒè¿‡ç¨‹ä¸­ä½¿ç”¨çš„æ˜¾å­˜ï¼ˆLoRA å¾®è°ƒéƒ¨åˆ†ï¼‰
used_percentage = round(used_memory / max_memory * 100, 3)  # å³°å€¼ä¿ç•™æ˜¾å­˜å æ€»æ˜¾å­˜çš„ç™¾åˆ†æ¯”
lora_percentage = round(used_memory_for_lora / max_memory * 100, 3)  # å¾®è°ƒæ˜¾å­˜å æ€»æ˜¾å­˜çš„ç™¾åˆ†æ¯”

# æ‰“å°è®­ç»ƒä½¿ç”¨çš„æ—¶é—´ï¼ˆç§’ï¼‰
print(f"{trainer_stats.metrics['train_runtime']} ç§’ç”¨äºè®­ç»ƒã€‚")

# æ‰“å°è®­ç»ƒä½¿ç”¨çš„æ—¶é—´ï¼ˆåˆ†é’Ÿï¼‰
print(f"{round(trainer_stats.metrics['train_runtime']/60, 2)} åˆ†é’Ÿç”¨äºè®­ç»ƒã€‚")

# æ‰“å°å³°å€¼ä¿ç•™æ˜¾å­˜ï¼ˆGBï¼‰
print(f"å³°å€¼ä¿ç•™æ˜¾å­˜ = {used_memory} GBã€‚")

# æ‰“å°å¾®è°ƒè¿‡ç¨‹ä¸­ä½¿ç”¨çš„æ˜¾å­˜ï¼ˆGBï¼‰
print(f"å¾®è°ƒè¿‡ç¨‹ä¸­ä½¿ç”¨çš„å³°å€¼æ˜¾å­˜ = {used_memory_for_lora} GBã€‚")

# æ‰“å°å³°å€¼ä¿ç•™æ˜¾å­˜å æ€»æ˜¾å­˜çš„ç™¾åˆ†æ¯”
print(f"å³°å€¼ä¿ç•™æ˜¾å­˜å æ€»æ˜¾å­˜çš„ç™¾åˆ†æ¯” = {used_percentage} %ã€‚")

# æ‰“å°å¾®è°ƒæ˜¾å­˜å æ€»æ˜¾å­˜çš„ç™¾åˆ†æ¯”
print(f"å¾®è°ƒè¿‡ç¨‹ä¸­æ˜¾å­˜å æ€»æ˜¾å­˜çš„ç™¾åˆ†æ¯” = {lora_percentage} %ã€‚")
```

<a name="Inference"></a>
### æ¨ç†

è®©æˆ‘ä»¬è¿è¡Œæ¨¡å‹å§ï¼ä½ å¯ä»¥æ›´æ”¹æŒ‡ä»¤å’Œè¾“å…¥å†…å®¹â€”â€”å°†è¾“å‡ºç•™ç©ºå³å¯ï¼

**[æ–°åŠŸèƒ½] åœ¨å…è´¹çš„ Colab ä¸­ä½“éªŒ Llama-3.1 8b Instruct çš„ 2 å€é€Ÿæ¨ç†ï¼Œ[ç‚¹å‡»è¿™é‡Œ](https://colab.research.google.com/github/unslothai/notebooks/blob/main/nb/Unsloth_Studio.ipynb)**

```python
# alpaca_prompt = ä»ä¸Šæ–¹å¤åˆ¶çš„æ¨¡æ¿
FastLanguageModel.for_inference(model)  # å¯ç”¨åŸç”Ÿçš„ 2 å€é€Ÿæ¨ç†æ¨¡å¼
inputs = tokenizer(
[
    alpaca_prompt.format(
        "ç»§ç»­æ–æ³¢é‚£å¥‘æ•°åˆ—ã€‚",  # æŒ‡ä»¤
        "1, 1, 2, 3, 5, 8",  # è¾“å…¥
        "",  # è¾“å‡º - ç•™ç©ºä»¥è¿›è¡Œç”Ÿæˆï¼
    )
], return_tensors="pt").to("cuda")  # å°†è¾“å…¥è½¬æ¢ä¸ºå¼ é‡å¹¶è½¬ç§»åˆ° GPU

outputs = model.generate(**inputs, max_new_tokens=64, use_cache=True)  # ç”Ÿæˆæœ€å¤š 64 ä¸ªæ–° token
tokenizer.batch_decode(outputs)  # è§£ç ç”Ÿæˆçš„è¾“å‡º
```

 You can also use a `TextStreamer` for continuous inference - so you can see the generation token by token, instead of waiting the whole time!

```python
# alpaca_prompt = ä»ä¸Šæ–¹å¤åˆ¶çš„æ¨¡æ¿
FastLanguageModel.for_inference(model)  # å¯ç”¨åŸç”Ÿçš„ 2 å€é€Ÿæ¨ç†æ¨¡å¼
inputs = tokenizer(
[
    alpaca_prompt.format(
        "ç»§ç»­æ–æ³¢é‚£å¥‘æ•°åˆ—ã€‚",  # æŒ‡ä»¤
        "1, 1, 2, 3, 5, 8",  # è¾“å…¥
        "",  # è¾“å‡º - ç•™ç©ºä»¥è¿›è¡Œç”Ÿæˆï¼
    )
], return_tensors="pt").to("cuda")  # å°†è¾“å…¥è½¬æ¢ä¸ºå¼ é‡å¹¶è½¬ç§»åˆ° GPU

from transformers import TextStreamer  # å¯¼å…¥ TextStreamer
text_streamer = TextStreamer(tokenizer)  # åˆ›å»ºä¸€ä¸ªæ–‡æœ¬æµå¯¹è±¡ï¼Œç”¨äºé€æ­¥æ˜¾ç¤ºç”Ÿæˆå†…å®¹
_ = model.generate(**inputs, streamer=text_streamer, max_new_tokens=128)  # ä½¿ç”¨æµå¼ç”Ÿæˆï¼Œæœ€å¤šç”Ÿæˆ 128 ä¸ªæ–° token
```

<a name="Save"></a>
### ä¿å­˜ä¸åŠ è½½å¾®è°ƒåçš„æ¨¡å‹

è¦å°†æœ€ç»ˆæ¨¡å‹ä¿å­˜ä¸º LoRA é€‚é…å™¨ï¼Œå¯ä»¥ä½¿ç”¨ Hugging Face çš„ `push_to_hub` æ–¹æ³•è¿›è¡Œåœ¨çº¿ä¿å­˜ï¼Œæˆ–è€…ä½¿ç”¨ `save_pretrained` æ–¹æ³•è¿›è¡Œæœ¬åœ°ä¿å­˜ã€‚

**[æ³¨æ„]** è¿™åªä¼šä¿å­˜ LoRA é€‚é…å™¨ï¼Œè€Œä¸ä¼šä¿å­˜å®Œæ•´æ¨¡å‹ã€‚å¦‚æœéœ€è¦ä¿å­˜ä¸º 16bit æˆ– GGUF æ ¼å¼ï¼Œè¯·å‘ä¸‹æ»šåŠ¨æŸ¥çœ‹ï¼

```python
model.save_pretrained("lora_model")  # æœ¬åœ°ä¿å­˜æ¨¡å‹
tokenizer.save_pretrained("lora_model")  # æœ¬åœ°ä¿å­˜åˆ†è¯å™¨
# model.push_to_hub("your_name/lora_model", token="...")  # åœ¨çº¿ä¿å­˜æ¨¡å‹åˆ° Hugging Face Hub
# tokenizer.push_to_hub("your_name/lora_model", token="...")  # åœ¨çº¿ä¿å­˜åˆ†è¯å™¨åˆ° Hugging Face Hub
```

ç°åœ¨ï¼Œå¦‚æœä½ æƒ³åŠ è½½æˆ‘ä»¬åˆšåˆšä¿å­˜çš„ LoRA é€‚é…å™¨è¿›è¡Œæ¨ç†ï¼Œåªéœ€å°† `False` è®¾ç½®ä¸º `True`ï¼š

```python
if False:  # å¦‚æœæ¡ä»¶ä¸º Trueï¼Œåˆ™æ‰§è¡Œä»¥ä¸‹ä»£ç ï¼ˆå½“å‰æ˜¯ Falseï¼Œä»£ç ä¸ä¼šè¿è¡Œï¼‰
    from unsloth import FastLanguageModel  # ä» unsloth åº“å¯¼å…¥ FastLanguageModel
    model, tokenizer = FastLanguageModel.from_pretrained(
        model_name="lora_model",  # ä½ ç”¨äºè®­ç»ƒçš„æ¨¡å‹åç§°
        max_seq_length=max_seq_length,  # æœ€å¤§åºåˆ—é•¿åº¦
        dtype=dtype,  # æ•°æ®ç±»å‹ï¼ˆä¾‹å¦‚ fp16ã€fp32 ç­‰ï¼‰
        load_in_4bit=load_in_4bit,  # æ˜¯å¦ä»¥ 4bit æ ¼å¼åŠ è½½æ¨¡å‹ï¼ˆèŠ‚çœæ˜¾å­˜ï¼‰
    )
    FastLanguageModel.for_inference(model)  # å¯ç”¨åŸç”Ÿçš„ 2 å€é€Ÿæ¨ç†æ¨¡å¼

# alpaca_prompt = ä½ å¿…é¡»ä»ä¸Šæ–¹å¤åˆ¶æ¨¡æ¿ï¼

inputs = tokenizer(
[
    alpaca_prompt.format(
        "å·´é»è‘—åçš„é«˜å¡”æ˜¯ä»€ä¹ˆï¼Ÿ",  # æŒ‡ä»¤
        "",  # è¾“å…¥
        "",  # è¾“å‡º - ç•™ç©ºä»¥ç”Ÿæˆå†…å®¹ï¼
    )
], return_tensors="pt").to("cuda")  # å°†è¾“å…¥è½¬æ¢ä¸º PyTorch å¼ é‡ï¼Œå¹¶ç§»åŠ¨åˆ° GPU

from transformers import TextStreamer  # ä» transformers åº“å¯¼å…¥ TextStreamer
text_streamer = TextStreamer(tokenizer)  # åˆ›å»ºä¸€ä¸ªæ–‡æœ¬æµå¯¹è±¡ï¼Œç”¨äºé€æ­¥æ˜¾ç¤ºç”Ÿæˆå†…å®¹
_ = model.generate(**inputs, streamer=text_streamer, max_new_tokens=128)  # ä½¿ç”¨æµå¼ç”Ÿæˆï¼Œæœ€å¤šç”Ÿæˆ 128 ä¸ªæ–° token
```

ä½ ä¹Ÿå¯ä»¥ä½¿ç”¨ Hugging Face çš„ `AutoModelForPeftCausalLM`ã€‚ä»…åœ¨æœªå®‰è£… `unsloth` çš„æƒ…å†µä¸‹ä½¿ç”¨æ­¤æ–¹æ³•ã€‚ä¸è¿‡ï¼Œè¿™å¯èƒ½ä¼šéå¸¸æ…¢ï¼Œå› ä¸ºä¸æ”¯æŒä¸‹è½½ `4bit` æ¨¡å‹ï¼Œè€Œä½¿ç”¨ Unsloth çš„**æ¨ç†é€Ÿåº¦å¿« 2 å€**ã€‚

```python
if False:  # å¦‚æœæ¡ä»¶ä¸º Trueï¼Œåˆ™æ‰§è¡Œä»¥ä¸‹ä»£ç ï¼ˆå½“å‰æ˜¯ Falseï¼Œä»£ç ä¸ä¼šè¿è¡Œï¼‰
    # æˆ‘å¼ºçƒˆä¸æ¨èä½¿ç”¨è¿™ç§æ–¹å¼ - å¦‚æœå¯èƒ½ï¼Œè¯·ä½¿ç”¨ Unsloth
    from peft import AutoPeftModelForCausalLM  # ä» peft åº“å¯¼å…¥ AutoPeftModelForCausalLM
    from transformers import AutoTokenizer  # ä» transformers åº“å¯¼å…¥ AutoTokenizer
    model = AutoPeftModelForCausalLM.from_pretrained(
        "lora_model",  # ä½ ç”¨äºè®­ç»ƒçš„æ¨¡å‹åç§°
        load_in_4bit=load_in_4bit,  # æ˜¯å¦ä»¥ 4bit æ ¼å¼åŠ è½½æ¨¡å‹ï¼ˆèŠ‚çœæ˜¾å­˜ï¼‰
    )
    tokenizer = AutoTokenizer.from_pretrained("lora_model")  # åŠ è½½åˆ†è¯å™¨
```
