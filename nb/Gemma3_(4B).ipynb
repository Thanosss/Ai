### Unsloth

`FastModel` 现在支持加载几乎所有类型的模型！这包括 **视觉模型** 和 **文本模型**！

````python
from unsloth import FastModel
import torch

# 定义支持 4-bit 动态量化的模型列表
fourbit_models = [
    # 4-bit 动态量化：提供更高的准确性并降低内存使用
    "unsloth/gemma-3-1b-it-unsloth-bnb-4bit",
    "unsloth/gemma-3-4b-it-unsloth-bnb-4bit",
    "unsloth/gemma-3-12b-it-unsloth-bnb-4bit",
    "unsloth/gemma-3-27b-it-unsloth-bnb-4bit",

    # 其他流行的模型
    "unsloth/Llama-3.1-8B",
    "unsloth/Llama-3.2-3B",
    "unsloth/Llama-3.3-70B",
    "unsloth/mistral-7b-instruct-v0.3",
    "unsloth/Phi-4",
] # 更多模型请访问 https://huggingface.co/unsloth

# 加载模型和分词器
model, tokenizer = FastModel.from_pretrained(
    model_name = "unsloth/gemma-3-4b-it",  # 指定模型名称
    max_seq_length = 2048,                 # 设置最大序列长度，用于支持长上下文
    load_in_4bit = True,                   # 启用 4-bit 量化以减少内存使用
    load_in_8bit = False,                  # 【新功能】启用 8-bit 量化，精度更高但内存占用增加 2 倍
    full_finetuning = False,               # 【新功能】支持完整微调
    # token = "hf_...",                    # 如果使用受限模型（gated models），需要提供访问令牌
)
````

我们现在添加了 LoRA 适配器，因此我们只需要更新少量参数即可！

````python
model = FastModel.get_peft_model(
    model,
    finetune_vision_layers     = False, # 关闭视觉层微调，仅针对文本任务！
    finetune_language_layers   = True,  # 应该保持开启，用于语言层微调！
    finetune_attention_modules = True,  # 开启注意力模块微调，适合 GRPO（全局稀疏优化）
    finetune_mlp_modules       = True,  # 始终保持开启，用于 MLP 模块微调！

    r = 8,           # r 值越大，精度越高，但可能导致过拟合
    lora_alpha = 8,  # 推荐 alpha 至少等于 r
    lora_dropout = 0, # 不使用 dropout
    bias = "none",    # 不添加偏置项
    random_state = 3407, # 随机种子，确保结果的可重复性
)
````

<a name="Data"></a>
### 数据准备

我们现在使用 `Gemma-3` 格式进行对话风格的微调。我们采用了 [Maxime Labonne 的 FineTome-100k 数据集](https://huggingface.co/datasets/mlabonne/FineTome-100k)，其风格与 ShareGPT 类似。`Gemma-3` 格式将多轮对话渲染为如下形式：

```
<bos><start_of_turn>user
Hello!<end_of_turn>
<start_of_turn>model
Hey there!<end_of_turn>
```

我们使用自定义的 `get_chat_template` 函数来获取正确的对话模板。目前支持的格式包括：`zephyr, chatml, mistral, llama, alpaca, vicuna, vicuna_old, phi3, llama3, phi4, qwen2.5, gemma3` 等。

````python
from unsloth.chat_templates import get_chat_template
tokenizer = get_chat_template(
    tokenizer,
    chat_template = "gemma-3",  # 指定使用 "gemma-3" 对话模板
)
````

````python
from datasets import load_dataset
dataset = load_dataset("mlabonne/FineTome-100k", split="train")  # 加载名为 "mlabonne/FineTome-100k" 的数据集，并选择训练集
````

我们现在使用 `standardize_data_formats` 来尝试将数据集转换为适合微调的正确格式！

````python
from unsloth.chat_templates import standardize_data_formats
dataset = standardize_data_formats(dataset)
````

让我们来看看第 100 行数据是什么样的！

````python
dataset[100]
````

我们现在需要将 `Gemma-3` 对话模板应用到这些对话中，并将结果保存到 `text` 字段中。

````python
def apply_chat_template(examples):
    texts = tokenizer.apply_chat_template(examples["conversations"])
    return { "text" : texts }
pass
dataset = dataset.map(apply_chat_template, batched = True)
````

让我们看看应用对话模板后的效果！注意，`Gemma-3` 默认会添加一个 `<bos>`！

````python
dataset[100]["text"]
````

<a name="Train"></a>
### 训练模型

现在让我们使用 Huggingface TRL 的 `SFTTrainer`！更多文档请参考：[TRL SFT 文档](https://huggingface.co/docs/trl/sft_trainer)。为了加快进度，我们只进行 60 步训练，但你可以将 `num_train_epochs=1` 设置为完整训练，并关闭 `max_steps=None`。

````python
from trl import SFTTrainer, SFTConfig

# 使用 Huggingface TRL 的 SFTTrainer 进行微调
trainer = SFTTrainer(
    model=model,  # 要微调的模型
    tokenizer=tokenizer,  # 模型对应的分词器
    train_dataset=dataset,  # 训练数据集
    eval_dataset=None,  # 可设置评估数据集（这里为 None）
    args=SFTConfig(  # 配置训练参数
        dataset_text_field="text",  # 数据集中包含文本的字段
        per_device_train_batch_size=2,  # 每个设备上的训练批次大小
        gradient_accumulation_steps=4,  # 使用梯度累积模拟更大的批次大小
        warmup_steps=5,  # 学习率预热的步数
        # num_train_epochs=1,  # 可设置为完整训练运行的训练轮数
        max_steps=30,  # 最大训练步数（这里设置为 30）
        learning_rate=2e-4,  # 学习率（长时间训练可降低为 2e-5）
        logging_steps=1,  # 日志记录的间隔步数
        optim="adamw_8bit",  # 优化器类型（支持 8-bit 模式以节省显存）
        weight_decay=0.01,  # 权重衰减
        lr_scheduler_type="linear",  # 学习率调度器类型（线性）
        seed=3407,  # 随机种子以确保结果可复现
        report_to="none",  # 日志报告工具（可设置为 WandB 等）
    ),
)
````

我们还使用了 Unsloth 的 `train_on_completions` 方法，仅对助手的输出进行训练，并忽略用户输入的损失。这有助于提高微调的准确性！

````python
from unsloth.chat_templates import train_on_responses_only
trainer = train_on_responses_only(
    trainer,
    instruction_part = "<start_of_turn>user\n",
    response_part = "<start_of_turn>model\n",
)
````

让我们验证指令部分的屏蔽是否已完成！再打印第 100 行数据：

````python
tokenizer.decode(trainer.train_dataset[100]["input_ids"])
````

现在让我们打印出被屏蔽的示例——你应该只会看到答案部分：

````python
tokenizer.decode([tokenizer.pad_token_id if x == -100 else x for x in trainer.train_dataset[100]["labels"]]).replace(tokenizer.pad_token, " ")
````

````python
# @title 显示当前的内存状态
gpu_stats = torch.cuda.get_device_properties(0)
start_gpu_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
max_memory = round(gpu_stats.total_memory / 1024 / 1024 / 1024, 3)
print(f"GPU = {gpu_stats.name}。最大内存 = {max_memory} GB。")
print(f"已保留 {start_gpu_memory} GB 的内存。")
````

让我们开始训练模型吧！如果需要恢复一次中断的训练，可以设置 `trainer.train(resume_from_checkpoint=True)`。

````python
trainer_stats = trainer.train()
````

````python
# @title 显示最终的内存和时间统计
used_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
used_memory_for_lora = round(used_memory - start_gpu_memory, 3)
used_percentage = round(used_memory / max_memory * 100, 3)
lora_percentage = round(used_memory_for_lora / max_memory * 100, 3)
print(f"训练耗时 = {trainer_stats.metrics['train_runtime']} 秒。")
print(f"训练耗时 = {round(trainer_stats.metrics['train_runtime']/60, 2)} 分钟。")
print(f"峰值保留内存 = {used_memory} GB。")
print(f"训练的峰值保留内存 = {used_memory_for_lora} GB。")
print(f"峰值保留内存占最大内存的比例 = {used_percentage} %。")
print(f"训练的峰值保留内存占最大内存的比例 = {lora_percentage} %。")
````

<a name="Inference"></a>
### 推理
让我们通过 Unsloth 原生推理运行模型吧！根据 `Gemma-3` 团队的建议，推理的推荐设置为：`temperature = 1.0, top_p = 0.95, top_k = 64`

````python
from unsloth.chat_templates import get_chat_template
# 使用 Unsloth 的聊天模板获取适配的 tokenizer
tokenizer = get_chat_template(
    tokenizer,
    chat_template="gemma-3",  # 指定聊天模板为 "gemma-3"
)

# 定义输入消息
messages = [{
    "role": "user",  # 用户角色
    "content": [{
        "type": "text",  # 消息类型为文本
        "text": "继续序列：1, 1, 2, 3, 5, 8,",  # 输入的序列内容
    }]
}]

# 将消息应用聊天模板，并添加生成提示
text = tokenizer.apply_chat_template(
    messages,
    add_generation_prompt=True,  # 必须添加生成提示以支持生成
)

# 使用模型生成输出
outputs = model.generate(
    **tokenizer([text], return_tensors="pt").to("cuda"),  # 将输入转换为张量并移动到 GPU
    max_new_tokens=64,  # 设置生成的最大新 token 数量，可增加以获得更长输出
    # Gemma-3 推荐的推理参数
    temperature=1.0, top_p=0.95, top_k=64,
)

# 解码生成的输出
tokenizer.batch_decode(outputs)
````

您还可以使用 `TextStreamer` 进行连续推理——这样您可以逐个令牌地查看生成结果，而不必一直等待完整输出！

````python
# 定义用户输入消息
messages = [{
    "role": "user",  # 用户角色
    "content": [{"type": "text", "text": "为什么天空是蓝色的？"}]  # 输入问题
}]

# 应用聊天模板并添加生成提示
text = tokenizer.apply_chat_template(
    messages,
    add_generation_prompt=True,  # 必须添加生成提示以支持生成
)

# 导入 TextStreamer 用于连续推理
from transformers import TextStreamer
_ = model.generate(
    **tokenizer([text], return_tensors="pt").to("cuda"),  # 将输入转换为张量并移动到 GPU
    max_new_tokens=64,  # 设置生成的最大新 token 数量，可增加以获得更长输出
    # Gemma-3 推荐的推理参数
    temperature=1.0, top_p=0.95, top_k=64,
    streamer=TextStreamer(tokenizer, skip_prompt=True),  # 使用 TextStreamer 进行连续推理，跳过提示部分
)
````

**[注意]** 这只会保存 LoRA 适配器，而不是完整模型。如果需要保存为 16bit 或 GGUF 格式，请向下滚动查看！

````python
# 将模型保存到本地
model.save_pretrained("gemma-3")  # 本地保存
tokenizer.save_pretrained("gemma-3")

# 将模型上传到 Huggingface Hub（在线保存）
# model.push_to_hub("HF_ACCOUNT/gemma-3", token="...")  # 在线保存模型
# tokenizer.push_to_hub("HF_ACCOUNT/gemma-3", token="...")  # 在线保存分词器
````

现在，如果你想加载我们刚刚保存的 LoRA 适配器进行推理，将 `False` 设置为 `True`：

````python
# 如果需要加载 LoRA 适配器进行推理，将 False 改为 True
if False:
    from unsloth import FastModel
    model, tokenizer = FastModel.from_pretrained(
        model_name="lora_model",  # 你用于训练的模型名称
        max_seq_length=2048,
        load_in_4bit=True,        # 加载为 4bit 量化模型
    )

# 构建用户消息
messages = [{
    "role": "user",
    "content": [{"type": "text", "text": "What is Gemma-3?"}]
}]
# 通过分词器应用对话模板
text = tokenizer.apply_chat_template(
    messages,
    add_generation_prompt=True,  # 必须添加生成提示以进行生成
)

# 使用 Huggingface 的 TextStreamer 进行推理
from transformers import TextStreamer
_ = model.generate(
    **tokenizer([text], return_tensors="pt").to("cuda"),
    max_new_tokens=64,  # 增加此值以生成更长的输出
    # 推荐的 Gemma-3 设置
    temperature=1.0, top_p=0.95, top_k=64,
    streamer=TextStreamer(tokenizer, skip_prompt=True),  # 跳过提示部分
)
````
